```python
# Load the dataset (.csv) using pandas package

CMS_calori_dataset = pd.read_csv('hgcal.csv')

# .head directive on the panda dataframe displays the first n-rows

CMS_calori_dataset.head(n = 10)
```

```python
# Convert the panda dataframe into numpy 2D array

CMS_calori_dataset_np = CMS_calori_dataset.to_numpy()

# The converted numpy array has the dimension of 420 (rows) x 8 (columns)

print(CMS_calori_dataset_np.shape)
```

```python
# Extract only x, y, z, eta, phi and energy columns from the dataset and stack them along column direction
# Name this new 2D array CMS_calori_dataset_np_sub.
# The array should have dimension 420 (rows) x 6 (columns)

# YOUR CODE HERE
# Check column names to be sure
print("Column names:", CMS_calori_dataset.columns.tolist())

# Extract columns  for x y z eta phi and energy
CMS_calori_dataset_np_sub = CMS_calori_dataset_np[:, 1:7]

# Verify the shape
print("Shape of subset:", CMS_calori_dataset_np_sub.shape)  # Should be (420, 6)
```

```python
# Create the scaling function

def scale_data(arr):
    # Calculate the mean for each column
    # axis=0 means calculate along rows (for each column)
    mean = np.mean(arr, axis=0)

    # Calculate the standard deviation for each column
    std = np.std(arr, axis=0)

    # Perform the scaling formula
    # (arr - mean) subtracts mean from each element
    # Then divide by std
    # Add tiny number to avoid division by zero
    epsilon = 1e-10
    scaled_data = (arr - mean) / (std + epsilon)

    #Return the scaled array
    return scaled_data
```

```python
# Test the function with CMS_calori_dataset_np_sub

# Apply scaling to subset
CMS_calori_dataset_np_sub_scaled = scale_data(CMS_calori_dataset_np_sub)

# Verify it worked - check the first column (x coordinate)
print("Original x column - mean:", np.mean(CMS_calori_dataset_np_sub[:, 0]))
print("Scaled x column - mean:", np.mean(CMS_calori_dataset_np_sub_scaled[:, 0]))
print("Original x column - std:", np.std(CMS_calori_dataset_np_sub[:, 0]))
print("Scaled x column - std:", np.std(CMS_calori_dataset_np_sub_scaled[:, 0]))
```

```python
# Confirm the data is scaled for 'x' column

plt.figure(figsize = (10, 5))

plt.hist(CMS_calori_dataset_np_sub_scaled[:, 0], bins = 20, facecolor = 'grey', edgecolor = 'black', linewidth = 2)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add proper x-label and y-label
# YOUR CODE HERE
#scaled x coordinates on x axis, amount of events for each x-variable
plt.xlabel('Scaled X Coordinate (standard deviations)', fontsize=14)
plt.ylabel('Number of events (frequency)', fontsize=14)

plt.show()
```
